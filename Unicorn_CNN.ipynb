# 1. Best Frequency band selection

select 10 frequency bands

```python
from mne.decoding import CSP
from sklearn.svm import SVC
from sklearn.model_selection import ShuffleSplit,StratifiedKFold ,cross_val_score, cross_val_predict
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from scipy import signal

def butter_bandpass(lowcut,highcut,fs,order):
    nyq = 0.5*fs
    low = lowcut/nyq
    high = highcut/nyq
    b,a = signal.butter(order,[low,high],'bandpass')
    return b,a

def butter_bandpass_filter(data, lowcut = 6, highcut = 30, order = 8):
    b,a = butter_bandpass(lowcut,highcut,250,order)
    y = signal.filtfilt(b,a,data,axis=2)
    return y

def discriminative_frequency_band_selection(data, label_1, label_2):
    Bh, Bl = 32, 6  # Initial upper and lower frequency limits
    A1, A2 = 0, 0  # Initial accuracy values

    # Step 1: Finding Bh
    while A1 >= A2:
        Bh -= 2
        A2 = A1        
        if(Bl >= Bh):
            break
        A1 = train_classifier(Bh, Bl, data, label_1, label_2)  # Design filter and train classifier using CSP


    Bh += 1  # Adjust Bh back
    A1 = train_classifier(Bh, Bl, data, label_1, label_2)

    if A1 >= A2:
        A2 = A1
    else:
        Bh = Bh + 1 
        A1 = A2

    # Step 2: Finding Bl
    while A1 >= A2:
        Bl += 2
        A2 = A1
        if(Bl >= Bh):
            break
        A1 = train_classifier(Bh, Bl, data, label_1, label_2)  # Design filter and train classifier using CSP

    Bl -= 1  # Adjust Bl back
    A1 = train_classifier(Bh, Bl, data, label_1, label_2)

    if A1 >= A2:
        A2 = A1
    else:
        Bh = Bh + 1 
        A1 = A2

    return Bh, Bl, A1

# Dummy functions for the classifier and accuracy computation
def train_classifier(Bh, Bl, data, label_1, label_2):

    init_data = data["Raw_Epoch"][np.where((data['label'] == label_1) | (data['label'] == label_2))]
    init_label = data['label'][np.where((data['label'] == label_1) | (data['label'] == label_2))]

    # init_data = data["Raw_Epoch"]
    # init_label = data['label']

    filtered_data = butter_bandpass_filter(init_data, lowcut= Bl, highcut= Bh)

    x_train, x_test, y_train, y_test = train_test_split(filtered_data, init_label, test_size=0.3, random_state = 47, stratify=init_label)

    csp = CSP(n_components = 8, reg=None, log=None, rank= 'info' , cov_est = 'epoch')
    csp.fit(x_train, y_train)

    x_train = csp.transform(x_train)
    x_test = csp.transform(x_test)

    # Initialize SVM with a linear kernel
    clf = SVC()

    param_grid = {
        'C':[0.01, 0.1, 1, 10 ,100],
        'kernel': ['linear', 'rbf', 'poly'],  # Example kernels
    }

    grid_search = GridSearchCV(clf, param_grid, cv=10, scoring='accuracy')

    grid_search.fit(x_train, y_train)
    y_pred = grid_search.predict(x_test)

    return accuracy_score(y_test, y_pred)


indices = [0, 1, 2, 3]
pairs = []
band_high = []
band_low = []
acc= []
class_name = ['left', 'right', 'non', 'feet']

# Nested loop to generate all pairs without reversing and without self-pairing
for i in range(len(indices)):
    for j in range(i + 1, len(indices)):
        pairs.append((indices[i], indices[j]))
        Bh, Bl, A1 = discriminative_frequency_band_selection(EEG_Epochs[target_data], label_1= indices[i], label_2 = indices[j])
        band_high.append(Bh)
        band_low.append(Bl)
        acc.append(A1)

for i in range(len(indices)):
    for j in range(i + 1, len(indices)):
        print(f"The selected frequency band of {class_name[i]} vs {class_name[j]} is: Bl = {band_low[j-1+i]}, Bh = {band_high[j-1+i]}, Acc = {acc[j-1+i]}" )
```

# CNN Model

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import keras
from keras.layers import Dropout
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix, classification_report

# Define the number of folds
num_folds = 5

# Prepare KFold cross-validation
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# Define lists to store evaluation results across folds
accuracy_per_fold = []
confusion_matrices = []
classification_reports = []

# Variables to keep track of the best model
best_model_weights = None
best_val_accuracy = 0
best_fold_number = -1

num_classes = 4

y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

# Iterate over each fold
for fold, (train_idx, val_idx) in enumerate(kf.split(train_for_cnn)):
    print(f"Training on fold {fold + 1}/{num_folds}")
    
    # Split the data into training and validation sets for this fold
    x_train_fold, x_val_fold = train_for_cnn[train_idx], train_for_cnn[val_idx]
    y_train_fold, y_val_fold = y_train[train_idx], y_train[val_idx]
    
    # Define the model architecture
    model = models.Sequential()
    
    # Input layer
    model.add(layers.InputLayer(input_shape=(train_for_cnn.shape[1], 3, 1)))
    
    # Convolutional layers
    model.add(layers.Conv2D(16, (9, 5), activation='relu', padding='same'))
    model.add(layers.Conv2D(16, (9, 5), activation='relu', padding='same'))
    model.add(layers.Conv2D(16, (9, 5), activation='relu', padding='same'))
    model.add(layers.Conv2D(16, (9, 5), activation='relu', padding='same'))
    
    # Max Pooling layer
    model.add(layers.MaxPooling2D((2, 2), strides=2))
    
    # Fully Connected (Dense) layer
    model.add(layers.Flatten())
    model.add(layers.Dense(640, activation='relu'))
    model.add(Dropout(0.5))
    model.add(layers.Dense(4, activation='softmax'))  # Output layer for 4 classes
    
    # Compile the model
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    checkpoint = tf.keras.callbacks.ModelCheckpoint('best_model.h5', 
                                                monitor='val_accuracy', 
                                                save_best_only=True, 
                                                mode='max', 
                                                verbose=0)
    
    # Train the model
    history = model.fit(x_train_fold, y_train_fold,
                        epochs=100,
                        batch_size=24,
                        validation_data=(x_val_fold, y_val_fold),
                        callbacks=[checkpoint],
                        verbose=1)
    
    best_model_infold = tf.keras.models.load_model('best_model.h5')

    # Evaluate the model on the validation data for this fold
    val_loss, val_acc = best_model_infold.evaluate(x_val_fold, y_val_fold)
    print(f"Fold {fold + 1} - Validation accuracy: {val_acc}")
    
    # Store the accuracy for this fold
    accuracy_per_fold.append(val_acc)
    
    # Check if this fold's model is the best so far
    if val_acc > best_val_accuracy:
        best_val_accuracy = val_acc
        best_fold_number = fold + 1  # Fold number (1-based index)
        best_model_weights = best_model_infold.get_weights()  # Save the weights of the best model


# After all folds are evaluated
print(f"\nBest fold: Fold {best_fold_number} with validation accuracy: {best_val_accuracy}")

# To use the best model's weights later:
final_model = models.Sequential()
# Define the same architecture as the original model
final_model.add(layers.InputLayer(input_shape=(train_for_cnn.shape[1], 3, 1)))
final_model.add(layers.Conv2D(16, (9, 5), activation='relu', padding='same'))
final_model.add(layers.Conv2D(16, (9, 5), activation='relu', padding='same'))
final_model.add(layers.Conv2D(16, (9, 5), activation='relu', padding='same'))
final_model.add(layers.Conv2D(16, (9, 5), activation='relu', padding='same'))
final_model.add(layers.MaxPooling2D((2, 2), strides=2))
final_model.add(layers.Flatten())
final_model.add(layers.Dense(640, activation='relu'))
final_model.add(Dropout(0.5))
final_model.add(layers.Dense(4, activation='softmax'))

# Compile the model
final_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])

# Load the best weights
final_model.set_weights(best_model_weights)

# You can now evaluate this model on the test data, make predictions, etc.
# Evaluate the best model on the test data
test_loss, test_acc = final_model.evaluate(test_for_cnn, y_test)

print(f"Best model test accuracy: {test_acc}")

# Make predictions with the best model
predictions = final_model.predict(test_for_cnn)

# Convert predictions and true labels from one-hot encoding to class labels
predicted_classes = np.argmax(predictions, axis=1)
true_classes = np.argmax(y_test, axis=1)

# Calculate the confusion matrix
conf_matrix = confusion_matrix(true_classes, predicted_classes)
print("Confusion Matrix:")
print(conf_matrix)

# Print a classification report for more detailed metrics
print("Classification Report:")
print(classification_report(true_classes, predicted_classes))
```
